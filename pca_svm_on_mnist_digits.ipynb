```python
labels = l.head(20000)
data = d.head(20000)

print("the shape of sample data = ", data.shape)
```

```python
from sklearn.preprocessing import StandardScaler
standardized_data = StandardScaler().fit_transform(data)
print(standardized_data.shape)
```

```python
#find the co-variance matrix 
sample_data = standardized_data

# matrix multiplication 
covar_matrix = np.matmul(sample_data.T , sample_data)

print ( "The shape of variance matrix = ", covar_matrix.shape)
```

```python
def pca_function(n_feature):
  values, vectors = eigh(covar_matrix, eigvals=(783-n_feature+1,783))
  print("Shape of eigen vectors = ",vectors.shape)
  vectors = vectors.T
  print("Updated shape of eigen vectors = ",vectors.shape)
  new_datatranspross = np.matmul(vectors, sample_data.T)
  print ("Resultanat new data coordinates' shape ", vectors.shape, "X", sample_data.T.shape," = ", new_datatranspross.shape)
  new_datatranspross = np.vstack((new_datatranspross, labels)).T
  return values, vectors, new_datatranspross
```

```python
a,b,new_datatrans_2n = pca_function(2)
```

```python
#@title Plot
def emPlot(X, title=None):
    x_min, x_max = np.min(X, 0), np.max(X, 0)
    X = (X - x_min) / (x_max - x_min)
    
    fig = plt.figure(figsize=(28,28))
    ax = fig.add_subplot(1, 1, 1)
    
    for i in range(15000):
        plt.text(X[i, 0], X[i, 1], str(labels[i]),
                 color=plt.cm.Set1(labels[i]),
                 fontdict={'family': 'monospace', 'size': 9})
        
    if hasattr(offsetbox, 'AnnotationBbox'):
        shown_images = np.array([[1., 1.]]) 
        for i in range(15000):
            dist = np.sum((X[i] - shown_images) ** 2, 1)
            
            if np.min(dist) < 4e-3:
                continue
            shown_images = np.r_[shown_images, [X[i]]]
            imagebox = offsetbox.AnnotationBbox(
                offsetbox.OffsetImage(d.iloc[i].values.reshape(28,28), cmap=plt.cm.gray_r),
                X[i])
            ax.add_artist(imagebox)
    plt.xticks([]), plt.yticks([])
    plt.xlabel('Principal Component X', fontsize=20)
    plt.ylabel('Principal Component X+1', fontsize=20)
    #plt.savefig("mnist_pca.png")
    plt.show()
    
emPlot(new_datatrans_2n[:, 0:2])
```

```python
def explained_variance_ratio(values):
  value_sorted  = values[::-1]
  eig_vals_total = sum(value_sorted)
  explained_variance = [(i / eig_vals_total)*100 for i in value_sorted]
  explained_variance = np.round(explained_variance, 2)
  cum_explained_variance = np.cumsum(explained_variance)
  #print('Explained variance: {}'.format(explained_variance))
  #print('Cumulative explained variance: {}'.format(cum_explained_variance))
  plt.plot(cum_explained_variance, '-o')
  plt.xticks(np.arange(0,800+1,100))
  plt.xlabel('Number of components')
  plt.ylabel('Cumulative explained variance');
  plt.show()
```

```python
e,f, new_datachoosen =  pca_function(20)
```

```python
from sklearn import svm,metrics
from sklearn.model_selection import train_test_split
```

```python
n = 20
```

```python
X_train,X_test,y_train,y_test = train_test_split(new_datachoosen[:,:n],new_datachoosen[:,n],test_size=0.3,shuffle=False)
```

```python
time_start=time.time() 

clf = svm.SVC(gamma=0.001)
clf.fit(X_train, y_train)
predicted = clf.predict(X_test) # predict

print("Classification report for classifier %s:\n%s\n"
      % (clf, metrics.classification_report(y_test, predicted)))

print("Accuracy={}".format(metrics.accuracy_score(y_test,predicted )))

# time used
time_end = time.time()
print('Time to classify: %0.2f minutes.' % ((time_end-time_start)/60))
```

```python
import seaborn as sns
# Using the confusion matrix derived from the RFC, we see below the boosted accuracy of the prediction

accp = accuracy_score(y_test, predicted)
cmp = confusion_matrix(y_test, predicted)

plt.figure(figsize=(10,10))
sns.heatmap(cmp, annot=True, linewidths=.25, square = True, cmap = 'Blues_r')
plt.ylabel('Actual label', fontsize=14)
plt.xlabel('Predicted label', fontsize=14)
all_sample_title = 'Accuracy Score: {0}'.format(accp)
plt.title(all_sample_title, size = 15)
# plt.savefig('pca_heatmap.png')
plt.show
```
